```python
import numpy as np
import scipy.linalg as la
import math
import random
import time
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import multiprocessing as mp
from multiprocessing import Pool
from functools import partial
import os

def Id(n):
	return np.eye(n)
def fit0(x,p,A,B):
	return A*p**x+B
def fit1(x,p,A):
	return A*p**x+0.5
'''def Nkron(*args):
	result = np.matrix('1.0')
	for item in args:
		result = np.kron(result,item)
	return result'''
def NormalizeState(State):
	return State/la.norm(State)
def density(state):
	state = np.matrix(state)
	A = np.dot(state,state.getH())
	return A
def seq_num(op):
	g = op[0]
	for i in range(1,len(op)):
		g = cliff_list[op[i]][g]
	return g
def rb_list(N): #produce rbsequence
	op = []	
	for s in np.arange(0,N):
		op.append(random.randint(0,23))
	n = seq_num(op)	
	op.append(inv_list[n])
	return op
'''def irb_list(N,targ):
	op = []	
	for s in np.arange(0,N):
		op.append(random.randint(0,23))
		op.append(targ)
	n = seq_num(op)	
	op.append(inv_list[n])	
	return op'''
def seq_list(type,N_e,m):
	seq = []
	if(type==1):
		for i in range(N_e):
			a = rb_list(m)
			seq.append(a)
	if(type==2):
		for i in range(N_e):
			a = irb_list(m,targ)
			seq.append(a)
	return seq
def H(x,y,ctrl_noise):
	e = np.random.normal(loc=0,scale=ctrl_noise) #gaussian noise
	h = x*(e+Omega)*X+y*(Omega+e)*Y
	return h
def U(H,angle):
	time = angle/(2*Omega)
	return la.expm(-1j*H*time)
def dephase(znoise):
	return Z*np.random.normal(loc=0,scale=znoise) #gaussian noise z
def primitive(n,noise):
	if(n==0):
		return Id(2)
	if(n==1):
		return U(H(1,0,noise)+dephase(detune),pi)    # x-gate
	if(n==2):
		return U(H(1,0,noise)+dephase(detune),pi/2)  # x/2 
	if(n==3):
		return U(H(1,0,noise)+dephase(detune),-pi/2) # -x/2 
	if(n==4):
		return U(H(0,1,noise)+dephase(detune),pi)    # y
	if(n==5):
		return U(H(0,1,noise)+dephase(detune),pi/2)  # y/2
	if(n==6):
		return U(H(0,1,noise)+dephase(detune),-pi/2)  #- y/2
def cliff_ele(n,noise):#noisy clifford element
	u = Id(2)	
	a = params['C'+str(n)]
	for i in a:
		u = np.dot(primitive(i,noise),u)
	return u
def expv(a,rho):#expect vaule
	a = np.matrix(a)
	rho = np.matrix(rho)
	e = np.trace(np.dot(a,rho))
	return e.real
def cliff_seq(op):#produce prob in same length
	state =Zero
	prob=[]
	for i in range(len(op)):
		state = np.dot(cliff_ele(op[i],noise),state)
		#print(op[i])
		#print(state)
	rho = density(state)
	prob=expv(P0,rho)
	return prob
def seq_with_mp(seqs):
	pool = mp.Pool()
	results = pool.map(cliff_seq,[seqs]*rep)
	pool.close()
	pool.join
	return results
def seq_with_mp_1(seqs):
	pool = mp.Pool()
	results = pool.map(cliff_seq,seqs)
	pool.close()
	pool.join
	return results
if __name__ == '__main__':

	start_time = time.time()


	Zero = np.matrix('1;0')
	One = np.matrix('0;1')

	P0 = np.dot(Zero,Zero.getH())
	P1 = np.dot(One,One.getH())

	pi = math.pi

	X = np.matrix('0 1; 1 0')
	Y = np.matrix('0 -1j; 1j 0')
	Z = np.matrix('1 0; 0 -1')

	cliff = np.load('cliff.npy')
	params = np.load('params.npy',allow_pickle=True).item()
	cliff_list = np.load('cliff_list.npy')
	inv_list = np.load('inv_list.npy')
	Omega = 285.7e3
	#omega0 = 9e9

	m = int(input('paper(0), 10 gap(max m)\n'))
	if(m==0):
		paper_length=[1,2,3,4,5,6,8,10,13,16,20,25,32,40,50,63,79,100,126,158,200,251,316,398,501, 631, 794, 1000, 1259, 1585, 1995, 2512, 3162]
	else:
		paper_length=[i for i in range(1,m,10)]

	N_e = int(input('# types\n'))
	re = int(input('#  rep\n'))
	ratio = float(input('x , y Noise Strength %\n'))
	detune = float(input('z noise Detune (kHz) \n'))
	global rep
	rep=re
	exp_type = int(input('(1) SRB (2) IRB\n'))
	save = str(input('Save data? \n'))	
	
	if(exp_type==2):
		targ = int(input('Target Gate\n'))

	if(exp_type==1):
		filename = 'data/rb_Ne'+str(N_e)+'_ratio'+str(ratio)+'_detune'+str(detune)+'.npy'
	if(exp_type==2):
		filename = 'data/irb_targ'+str(targ)+'_Ne'+str(N_e)+'_ratio'+str(ratio)+'_detune'+str(detune)+'.npy'
	prob={}
	detune = detune*1e3
	noise = Omega*ratio
	print('==================================')
	print(f'Detune {detune} Hz')
	print(f'Noise strength {noise} Hz')
	print('==================================')
	if(rep==1):
		p = []
		err = []
		for i in paper_length:
			op_list = seq_list(exp_type,N_e,i)
			result = seq_with_mp_1(op_list)
			prob['p'+str(i)] = result
			p.append(np.mean(result))
			err.append(np.std(result))
			print(f'Finished (length {i})')
			#print(f'Average Survival Probability {mu}')
	else:
		p = []
		err = []
		for i in paper_length:
			op_list = seq_list(exp_type,N_e,i)
			result = []
			for j in range(N_e):
				#print(f'Finished (type {j})')
				result.append(seq_with_mp(op_list[j]))
				#print('res=',result)
			#prob['p'+str(i)] = np.mean(result)
			p.append(np.mean(result))
			err.append(np.std(result))
			print(f'Finished (length {i})')
			#print(f'Average Survival Probability {mu}')
	if(save=='y'):
		np.save(filename,prob)
	print(p)
	print(err)    
	mean, covar = curve_fit(fit0,paper_length,p,p0=[1,1,1],bounds=(0,1),maxfev=10000)
	print('OLS Fit [r,p,A,B]','%.3g' % ((1-mean[0])*0.5),'%.3g' % mean[0],'%.3g' % mean[1],'%.3g' % mean[2])
	mean1, covar1 = curve_fit(fit1,paper_length,p,p0=[1,1],bounds=(0,1),maxfev=10000)
	print('OLS & B=0.5 Fit [r,p,A]','%.3g' % ((1-mean1[0])*0.5),'%.3g' % mean1[0],'%.3g' % mean1[1])
	mean2, covar2 = curve_fit(fit1,paper_length,p,p0=[1,1],sigma=err,absolute_sigma = True,bounds=(0,1))
	print('WLS & B=0.5 Fit [r,p,A]','%.3g' % ((1-mean2[0])*0.5),'%.3g' % mean2[0],'%.3g' % mean2[1])
	fig = plt.figure()
	ax = fig.add_subplot(111)
	ax.plot(paper_length,p,'.',color='r',markerfacecolor='none',markersize=9,label='RB data')
	ax.plot(paper_length,fit0(paper_length, *mean),'-',color='r',label='OLS')
	ax.plot(paper_length,fit1(paper_length, *mean1),'-',color='b',label='OLS&B=0.5')
	ax.plot(paper_length,fit1(paper_length, *mean2),'-',color='g',label='WLS&B=0.5')
	#ax.errorbar(length,p,color='r',yerr=err,capsize=8,fmt=' ')	
	ax.set_xscale('log')
	ax.legend(loc='best')
	plt.savefig("filename.png")
	plt.show()
```
